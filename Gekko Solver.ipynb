```python
import numpy as np
from gekko import GEKKO
import matplotlib.pyplot as plt
from opt_solver_util import (make_assignment_visual, 
                             solver_results, 
                             init_solver,
                             get_makespan
                            )
```

```python
# define dependencies; g[0] corresponds to dependencies of task 0 
# i.e. g[0] / s[0] + s[0] is T + E of task 0
g = [1, 2, 1, 3, 5]

m, s, O = init_solver(g)

# Manually add constraints
m.Equation(1 / s[2] == 1 / s[1] +  1 / s[3]) 

s, task_process_time = solver_results(s, m, O, verbose=True)

# Comment out code starting here if you don't want visual

# Manually define idle time dependencies if necessary
# Used to help make visual
# Be careful to not copy/paste if not needed
t = [[None, None] for _ in range(len(g))]
t[2][0] = task_process_time[4]

# define tasks assignment to each machine
machine_task_list = [[2], [4, 3, 1, 0]]

make_assignment_visual(t, task_process_time, machine_task_list)
print("makespan: " + str(get_makespan(task_process_time, machine_task_list)))
```
