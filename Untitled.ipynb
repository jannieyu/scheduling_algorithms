```python
import numpy as np
```

```python
def function(m1, m2):
    a = 2 * sum([2* np.sqrt(i) for i in range(1, min(m1, m2) + 1)])
    if (m1 != m2):
        a += sum([2*np.sqrt(j) for j in range(min(m1, m2) + 1, max(m1, m2) + 1)])
        
    return a
```

```python
t_values = np.arange(10, 10000, 10)
lst = []

for t in t_values:

    half_t = t/2


    optimal =  sum([4 * np.sqrt(i) for i in range(int(half_t) + 1)])
    suboptimal =  sum([2 * np.sqrt(i) for i in range(t + 1)])
    ratio = suboptimal/optimal
    print(t)
    print(ratio)
    lst.append(ratio)
```
