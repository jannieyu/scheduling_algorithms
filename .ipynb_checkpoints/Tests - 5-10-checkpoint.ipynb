```python
from gurobipy import *
import networkx as nx
import numpy as np
import random
import math
import matplotlib.pyplot as plt
from simulations import Test
from simulations_old import Test_old
from getf import GETF
from opt_solver_util import (make_assignment_visual, 
                             make_graph_visual,
                              solver_results, 
                              init_solver,
                              get_makespan,
                            v_helper,
                            define_constraints,
                            add_constraints)
import networkx as nx
```

```python
save_lst = []

for b in range(10):
    
    max_ratio = 0
    max_save = None
    
    
    num_tasks = 10
    num_machines = 3
    num_edges = 10
    tie_breaking_rule = 1
    simulations = Test(num_tasks, num_machines, num_edges, tie_breaking_rule, random=True)
    v, dependencies = v_helper(simulations.G, num_tasks, simulations.task_prev, simulations.machine_task_list, simulations.t)    
    m, s, O = init_solver(v)
    equality_constraints, relaxing_constraints = define_constraints(simulations.G, simulations.time_chunks, dependencies)
    m, s = add_constraints(m, s, equality_constraints, relaxing_constraints)



    critical_path = nx.dag_longest_path_length(simulations.G)
#     print("hre" + str(critical_path))
    critical_path_cost = sum([2 * np.sqrt(i) for i in range(1, critical_path + 2)])
    print([2 * np.sqrt(i) for i in range(1, critical_path + 2)])
#     print("hre" + str(critical_path + 1))
    average_path_cost =  num_machines * sum([2 * np.sqrt(i) for i in range(1 + int(np.floor(num_tasks/num_machines)))])
    worst_cost = sum([2 * np.sqrt(i) for i in range(num_tasks + 1)])
    s, task_process_time, obj = solver_results(s, m, O, verbose=True)



    crit_ratio = obj / critical_path_cost
    avg_ratio = obj / average_path_cost 
    ratio_lst = [avg_ratio, crit_ratio]
#     print("critical path cost = " + str(critical_path_cost))
#     print("average path cost = " + str(average_path_cost))
#     print("worst cost = " + str(worst_cost))
#     print("obj / avg ratio = " + str(avg_ratio))
#     print("obj / crit ratio = " + str(crit_ratio))
    make_assignment_visual(task_process_time, dependencies, simulations.machine_task_list)
    make_graph_visual(simulations.G, simulations.n)
    print("_________________________________")
#     if max(crit_ratio, avg_ratio) > max_ratio:
#         max_ratio = max(crit_ratio, avg_ratio)
#         max_save = max_ratio, simulations, ratio_lst, dependencies, task_process_time, x
# save_lst.append(max_save)
# print(x)
```
