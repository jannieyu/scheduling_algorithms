```python
from gurobipy import *
import networkx as nx
import numpy as np
import random
import math
import matplotlib.pyplot as plt
from simulations import Test
from simulations_old import Test_old
from getf import GETF
from opt_solver_util import (make_assignment_visual, 
                             make_graph_visual,
                              solver_results, 
                              init_solver,
                              get_makespan,
                            v_helper,
                            define_constraints,
                            add_constraints)
import networkx as nx
```

```python
num_tasks = 9
num_machines = 2
task_units = [1 for _ in range(num_tasks)]
task_transfer_units = {(0,2): 1, (2,3):1, (3,5):1, (2,4):1, (4,5):1, (5,6):1, (6,7):1,  (7, 8):1}
machine_speeds = [1 for _ in range(num_machines)]
machine_comm_bandwidths = [[-1 for _ in range(num_machines)] for _ in range(num_machines)]
tie_breaking_rule = 1
    
simulations = Test_old(task_units, \
    task_transfer_units, \
    machine_speeds, \
    machine_comm_bandwidths, \
    tie_breaking_rule)

v, dependencies = v_helper(simulations.G, num_tasks, simulations.task_prev, simulations.machine_task_list, simulations.t)    
m, s, O = init_solver(v)
equality_constraints, relaxing_constraints = define_constraints(simulations.G, simulations.time_chunks, dependencies)
m, s = add_constraints(m, s, equality_constraints, relaxing_constraints)
s, task_process_time, obj = solver_results(s, m, O, verbose=True)
make_assignment_visual(task_process_time, dependencies, simulations.machine_task_list)
```
